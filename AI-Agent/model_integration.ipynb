```python
from langchain.chat_models import init_chat_model
```

```python
model = init_chat_model(
    model='llama3.1',
    model_provider='ollama'
)
model
```

```python
model.invoke('hello, how are you?').content
```

```python
for chunk in model.stream('write a 100 word paragraph on ai'):
    print(chunk.text, end='|', flush=True)
```

```python
responses = model.batch([
    'why do parrots have beautiful colours?',
    'what is a markov chain',
    'where is the biggest black hole located?'
    ])

for response in responses:
    print(response.content + '\n\n---------------------------\n\n')
```
